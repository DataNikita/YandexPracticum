### Продажи игр за каждый год в зависимости от платформы
Посмотрим, как менялись продажи по платформам. Выберем платформы с наибольшими суммарными продажами и построим распределение по годам. Определим, за какой характерный срок появляются новые и исчезают старые платформы.  
Чтобы посмотреть, как менялись продажи по платформам, на основе таблицы `games` создадим сводную таблицу `platform`, которая будет отражать продажи каждой платформы по годам.

```python
platform = games.pivot_table(index = ['platform'], values = 'total_sales', columns = 'year_of_release', aggfunc = 'sum')
```

Выведем первые 5 строк сводной таблицы `platform`, чтобы убедиться, что она была сформирована корректно

```python
platform.head()
```

На основе сводной таблицы `platform` создадим тепловую карту с помощью метода `heatmap()`

```python
ax = plt.subplots(figsize = (18, 9))
sea.heatmap(platform,
            cmap = 'RdYlGn',
            linewidths = 0.1,
            )
plt.title('Продажи на платформах по годам', fontsize = 15)
plt.xlabel('Год', fontsize = 12)
plt.xticks(rotation=30, fontsize = 12)
plt.yticks(fontsize = 12)
plt.ylabel('Игровая платформа', fontsize = 15);
```

Для того, чтобы понять, за какой характерный срок появляются новые и исчезают старые платформы, создадим ещё одну сводную таблицу `time_of_platform` с помощью метода `groupby()`. Сделаем мы это в несколько действий:
1. Сначала создадим таблицу, в которой для каждой платформы будут отображены её суммарные продажи и год, в который была сделана первая продажа на этой платформе.

```python
platform_year_min = games.groupby(['platform']).agg({ 'total_sales' : 'sum', 'year_of_release' : 'min'}).reset_index()
platform_year_min = platform_year_min.rename(columns = {'year_of_release' : 'year_of_release_min'})
#platform_year_min
```

2. Создадим ещё одну таблицу, в которой для каждой платформы будет указан год, в который была сделана последняя продажа на этой платформе.

```python
platform_year_max = games.groupby(['platform']).agg({'year_of_release' : 'max'}).reset_index()
platform_year_max = platform_year_max.rename(columns = {'year_of_release' : 'year_of_release_max'})
#platform_year_max
```

3. Объединим таблицы `platform_year_min` и `platform_year_max` по столбцу `platform`, получив таблицу `time_of_platform`.

```python
time_of_platform = platform_year_min.merge(platform_year_max, on='platform', how='left')
#time_of_platform
```

4. Обратим внимание на год релиза для платформы `DS`. В таблице приведен 1985 год. Из тепловой карты, построенной в выше можно заметить, что продажи на данной платформе были сделаны впервые в 1985 году, а далее продажи возобновились только в 2004 году. Получившийся промежуток времени в 19 лет может исказить дальнейшие вычисления. Поэтому для данной платформы мы будем считать, что продажи начались в 2004 году. В получившейся таблице для платформы `DS` заменим значение 1985 на 2004 в столбце `year_of_release_min` 

```python
time_of_platform.loc[time_of_platform['platform'] == 'DS', 'year_of_release_min'] = 2004
```

5. Далее для того, чтобы получить промежуток времени, в течении которого платформа была "жива", отнимем от столбца `year_of_release_max` столбец `year_of_release_min`, тем самым получив столбец `life_of_platform` со сроком жизни платформы.

```python
time_of_platform['life_of_platform'] = time_of_platform['year_of_release_max'] - time_of_platform['year_of_release_min']
#time_of_platform
```

6. Т.к. нас интересуют платформы с наибольшим количеством продаж, то отсортируем получившуюся таблицу с помощью метода `sort_values()` по столбцу `total_sales` и выведем получившуюся таблицу.

```python
time_of_platform = time_of_platform.sort_values(by = 'total_sales', ascending = False).reset_index(drop = True)
time_of_platform
```

7. Воспользуемся методом `describe()`, чтобы узнать за какой характерный срок появляются новые и исчезают старые платформы

```python
time_of_platform['life_of_platform'].describe()
```

Из результатов метода `describe()` заметим, что в среднем, "жизнь" платформы составляет 7 лет, а медианное значение равно 6 годам. Исходя из данных замечаний, сделаем вывод, что характерным сроком за который появляются новые и исчезают старые платформы можно назвать срок 7 лет.

## Проверка гипотез

### Гипотеза №1 - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Проверим гипотезу, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Для начала сформируем нулевую (Но) и альтернативную (Н1) гипотезы:
* Hо - пользовательские рейтинги платформ Xbox One и PC одинаковые;
* H1 - пользовательские рейтинги платформ Xbox One и PC разные.

Гипотезы были сформулированы в таком виде, потому что при из формулировке, нулевая гипотеза Hо часто формулируется так, чтобы использовать знак равенства. Т.е. необходимо сделать предположение, что среднее значение определенной генеральной совокупности равно конкретному значению. В нашем случае конкретным значением стало среднее значение другой генеральной совокупности. Альтернативная гипотеза подразумевает, что средние значения нашей генеральной совокумности Не будет равно конкретному значению.

Для проверки гипотезы, возьмем две группы данных по каждой из платформ - `games_actual_PC_user` и `games_actual_XOne_user`\
Первая группа `games_actual_PC_user` - отдельный датафрейм по платформе PC, созданный на этапе выполнения п.3.7.;\
Вторая группа `games_actual_XOne_user` - отдельный датафрейм по платформе XOne, созданный на этапе выполнения п.3.7.

Из указанных датафреймов уже убраны "заглушки", поэтому дополнительно обрабатывать их не надо.

```python
games_actual_PC_user.head()
```

```python
games_actual_XOne_user.head()
```

Так как альтернативная гипотеза подразумевает то, что среднее выбранной генеральной совокупности НЕ БУДЕТ равно конкретному значению, что тут необходимо понять, КАК в нашем случае оно будет не равно. Т.е. оно будет больше или меньше нашего значения. Если больше, то это правосторонняя гипотеза. Если меньше, то левосторонняя. Если нет разницы больше или меньше и главное, чтобы было не равно, то это будет двусторонняя гипотеза.

В данном случае необходимо проверить двустороннюю гипотезу о равенстве среднего двух генеральных совокупностей. Для этого воспользуемся методом `scipy.stats.ttest_ind()`. Он подходит для проверки двух генеральных совокупностей между собой, потому что данные генеральные совокупности не зависят друг от друга и их выборочные средние нормально распределены.

```python
# первая группа - столбец с оценкой пользователей из датафрейма, где собраны игры только на PC
games_actual_PC_user['user_score'] 
# вторая группа - столбец с оценкой пользователей из датафрейма, где собраны игры только на XOne
games_actual_XOne_user['user_score'] 
alpha_1 = 0.05 # уровень статистической значимости
# если p-value окажется меньше него, отвергнем гипотезу

results_1 = st.ttest_ind(games_actual_PC_user['user_score'], games_actual_XOne_user['user_score'], equal_var = False)
# В данном случае будем считать, что их дисперсии взятых выборок не равны. 
# Поэтому параметр equal_var задаем равным False.
# Так как мы будем проверять двустороннюю гипотезу, то параметр alternative не будем задавать. Тогда он автоматически
# будет равен значению по-умолчанию two-sided (двустороняя)

print('p-значение:', results_1.pvalue)
if results_1.pvalue < alpha_1:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Промежуточный вывод:**  

Полученное значение p-value больше заданного уровня значимости, поэтому нельзя отвергнуть нулевую гипотезу.
Это значит, что пользовательские рейтинги платформ Xbox One и PC разные. Проверим это, выведя средние значения пользовательских рейтингов.

[Вернуться в начало](#Содержание)

```python
print ('Среднее значение пользовательского рейтинга PC равно', games_actual_PC_user['user_score'].mean())
print ('Среднее значение пользовательского рейтинга XOne равно', games_actual_XOne_user['user_score'].mean())
if (games_actual_PC_user['user_score'].mean() == games_actual_XOne_user['user_score'].mean()) :
    print ('Рейтинги равны')
else:
    print ('Рейтинги не равны')
```

Для начала сформируем нулевую (Но) и альтернативную (Н1) гипотезы:
* Hо - пользовательские рейтинги жанров Action и Sports равны;
* H1 - пользовательские рейтинги жанров Action и Sports не равны.

Для проверки гипотезы, создадим две группы данных по каждой из платформ - `games_actual_action` и `games_actual_sports`\
Первая группа `games_actual_action` - отдельный датафрейм по жанру Action;\
Вторая группа `games_actual_sports` - отдельный датафрейм по платформе Sports.

При создании указанных групп, также исключим игры, для которых оценка пользователей равна 0

```python
games_actual_action = games_actual.query('genre == "Action" and user_score != -5').reset_index(drop = True)
#games_actual_action.head()
```

```python
games_actual_sports = games_actual.query('genre == "Sports" and user_score != -5').reset_index(drop = True)
#games_actual_sports.head()
```

В данном случае необходимо проверить двустороннюю гипотезу о равенстве среднего двух генеральных совокупностей. Для этого воспользуемся методом `scipy.stats.ttest_ind()`.

```python
# первая группа - столбец с оценкой пользователей из датафрейма, где собраны игры жанра action
games_actual_action['user_score']
# вторая группа - столбец с оценкой пользователей из датафрейма, где собраны игры жанра sports
games_actual_sports['user_score'] 

alpha_1 = 0.05 # уровень статистической значимости
# если p-value окажется меньше него, отвергнем гипотезу

results_1 = st.ttest_ind(games_actual_action['user_score'], games_actual_sports['user_score'], equal_var = False)
# В данном случае будем считать, что их дисперсии взятых выборок не равны. 
# Поэтому параметр equal_var задаем равным False.
# Так как мы будем проверять двустороннюю гипотезу, то параметр alternative не будем задавать. Тогда он автоматически
# будет равен значению по-умолчанию two-sided (двустороняя)

print('p-значение:', results_1.pvalue)
if results_1.pvalue < alpha_1:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Промежуточный вывод:**  

Полученное значение p-value меньше заданного уровня значимости. Поэтому, на имеющихся данных и на уровне значимости 5% есть основания отвергнуть нулевую гипотезу о равенстве средних значений пользовательских рейтингов (отзывов) жанров Action и Sports в пользу альтернативной гипотезы.
