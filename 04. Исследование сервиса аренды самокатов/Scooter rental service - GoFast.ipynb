##### Соотношение пользователей с подпиской и без

```python
(users[users['subscription_type'] == 'free']['subscription_type'].count() * 100 / 
    users['subscription_type'].count() )
```

#### Проверка гипотез

##### Гипотеза №1 - Тратят ли пользователи с подпиской больше времени на поездки

Важно понять, тратят ли пользователи с подпиской больше времени на поездки? Если да, то пользователи с подпиской могут быть «выгоднее» для компании. Проверим гипотезу. Используем исходные данные о продолжительности каждой сессии — отдельно для подписчиков и тех, у кого нет подписки.

Т.к. объем данных не слишком большой, то не будем делать отдельную выборку и возьмем все данные.\
Для начала сформируем нулевую (Но) и альтернативную (Н1) гипотезы:
* Hо - пользователи с подпиской и без тратят одинаковое количество времени на поездки.
* H1 - пользователи с подпиской тратят больше времени на поездки.

Для проверки гипотезы, возьмем две группы пользователей - с подпиской и без.\
Первой группой является столбец `duration` из датафрейма c данными о пользователях с подпиской.\
Второй группой является  аналогичный столбец, но из датафрейма c данными о пользователях без подписки.

В данном случае необходимо проверить одностороннюю (правостороннюю) гипотезу о равенстве среднего двух генеральных совокупностей. Для этого воспользуемся методом scipy.stats.ttest_ind().

```python
# первая группа - столбец с длительностью поездок из датафрейма, где пользователи с подпиской
data_general_ultra['duration'] 

# вторая группа - столбец с длительностью поездок из датафрейма, где пользователи без подписки
data_general_free['duration'] 

alpha_1 = 0.05 # уровень статистической значимости
# если p-value окажется меньше него, отвергнем гипотезу

results_1 = st.ttest_ind(data_general_ultra['duration'], data_general_free['duration'], alternative = 'greater')
# Т.к. выборки достаточно большие, будем принимать, что их дисперсии равны. 
# А так же, т.к. мы сравниваем средние двух генеральных совокупностей, 
# то нулевую гипотезу можно сформулировать так: «средние значения совокупностей равны».
# Поэтому параметр equal_var оставляем по умолчанию, равным True.
print('p-значение:', results_1.pvalue)
if results_1.pvalue < alpha_1:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Промежуточный вывод:**  

Полученное значение p-value практически равно нулю.  
Столь малое значение p-value говорит о том, что о с 0% вероятности мы можем получить наблюдаемые данные, при условии, что нулевая гипотеза верна.
Т.к. полученное значение p-value меньше заданного уровня значимости, то мы можем отвергнуть нулевую гипотезу.\
Есть основания говорить, что пользователи с подпиской тратят больше времени на поездки.\

Расстояние одной поездки в 3130 метров — оптимальное с точки зрения износа самоката. Можно ли сказать, что среднее расстояние, которое проезжают пользователи с подпиской за одну поездку, не превышает 3130 метров? Проверьте гипотезу и сделайте выводы.

Т.к. объем данных не слишком большой, то не будем делать отдельную выборку и возьмем все данные.\
Для начала сформируем нулевую (Но) и альтернативную (Н1) гипотезы:

* Hо - пользователи с подпиской в среднем проезжают расстояние, равное 3130 метрам.
* H1 - пользователи с подпиской в среднем проезжают расстояние, больше 3130 метров.

Для проверки гипотезы возьмем столбец `distance` из датафрейма c данными о пользователях с подпиской.\

В данном случае необходимо проверить одностороннюю (правостороннюю) гипотезу. Для этого воспользуемся методом scipy.stats.ttest_1samp().

```python
data_general_ultra['distance'] # столбец с расстоянием, которое проезжают пользователи с подпиской

interested_value = 3130 # расстояние, оптимальное с точки зрения износа самоката

alpha = 0.05 # уровень статистической значимости

results_2 = st.ttest_1samp(data_general_ultra['distance'], interested_value, alternative = 'greater')
# т.к. мы проверяем одностороннюю гипотезу, то параметр alternative принимаем равным greater.

print('p-значение:', results_2.pvalue)

if results_2.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Ещё более отличительным фактором является то, что Кира не одна такая и пользователей, которые пользуются услугами сервиса зимой - много. Возможно стоит проверить способ сбора даных. Может оказаться так, что Кира родом из Тюмени, но живет в Сочи, где зимой температура отпускается днём до около 10 градусов выше нуля, что позволяет ей пользоваться услугами сервиса.
2. Из визуализации, проведенной в п.4.3. видно, что у пользователей без подписки есть поездки длительностью 30 секунд и на расстояние 0.85 метра. Возможно следует удалить из датафрейма поездки, длительность которых менее 5 минут и расстояние поездки менее, чем 10 метров. Скорее всего это "ошибочные" поездки, когда пользователь передумал ехать или имел другие причины завершить только что начатую поездку. Такие поездки искажают общую картину. Возможно, после удаления таких поездок, данные о пользователях без подписки, примут иной вид, что может повлиять на результаты исследования.
