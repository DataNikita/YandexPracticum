#### Обработка пропущенных значений

С помощью метода `isna()` выведем количество пропущенных значений в каждом из столбцов датасета. Также дополнительно выведем таблицу, показывающую долю пропущенных значений об общего кол-ва строк в столбце.

```python
display(pd.DataFrame(round(banks.isna().sum(),1)).style.background_gradient('coolwarm'))
pd.DataFrame(round(banks.isna().mean()*100,1)).style.background_gradient('coolwarm')
```

В столбце `age` обнаружено всего 26 пропущеных значений, что составляет 0,3 % от всех строк датасета. Между этими пропусками и данными в датасете нет зависимости и скорее всего пропуски появились случайно. Исходя из этого их можно отнести к типу MCAR(missing completely at random - отсутствует совершенно случайно). Удаление этих пропусков не исказит предоставленные данные и не повлияет на результаты исследования, поэтому удалим пропущенные значения в столбце `age`.

```python
banks = banks.dropna(subset = ['age'])
print('Количество строк датасета после удаления пропусков в столбце age равно:', banks.shape[0])
```

Как мы видим из результата выше, было удалено 26 строк.

В столбце `balance` обнаружено 2295 пропусков, что составляет 23 % от всех строк датасета. Для принятия решения по обработке пропущенных значений в этом столбце, а также принятия решения об их классификации, необходимо провести анализ связи пропусков с параметрами таблицы.

Для обработки пропущенных значений в столбце `balance` добавим временный столбец `skip_balance` в котором будет указано `'1'` при наличии пропуска и `'0'` при его отсутствии.

```python
banks['skip_balance'] = banks['balance'].isna()
banks['skip_balance'] = banks['skip_balance'].apply(lambda x: 1 if x == True else 0)
```

Проверим корректность выполнения команды по добавлению временного столбца `skip_balance`

```python
banks.head()
```

Добавление столбца прошло корректно.

Построим таблицу, которая отразит корреляцию полученного столбца `skip_balance` с остальными столбцами датасета. Это поможет нам понять есть ли зависимость между пропусками и другими данными в целом. 

```python
plt.subplots(figsize=(18, 8))
sns.heatmap(banks.corr()[['skip_balance']].sort_values(by = 'skip_balance', ascending=False, key=abs),
            annot = True,
            linewidths = 1,
            cmap = 'coolwarm',
            center= 0)

plt.title('Корреляция пропусков с остальными параметрами', fontsize = 15)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.show();
```

**Из построенной таблицы можно сделать вывод, что пропуски в столбце `balance` сильнее всего зависят от данных в столбцах `equity` и `products`.**

* Аналогично со столбцом `products` - для клиентов "с пропуском" видно, что они используют в основном только 1 продукт банка, редко когда 2 и больше 3 не использует никто. А для клиентов "без пропуска", характерно использование в основном 2 продуктов банка и встречаются клиенты, которые пользуются 4 и 5 продуктами.

Исходя из вышеперечисленного, а также принимая во внимание ранее построенную таблицу корреляции, можно сделать вывод, что значения в столбце `balance` имеют имеют зависимость от значений столбцов `equity` и `balance`, а это значит, что мы можем отнести их к типу MNAR (Missing Not At Random / Отсутствует не случайно). Это могут быть как и новые клиенты банка, которые еще не успели начать пользоваться другими предложениями банка, так и люди, которые открыли тот или иной продукт под определенную задачу и сразу снимают деньги при их поступлении.

```python
typical_user = round( banks.groupby(['churn']).agg({ 'score': {'mean', 'median'},
                                                #'city': pd.Series.mode,
                                                #'gender': pd.Series.mode,
                                                'age': {'mean', 'median'},
                                                'equity': 'median',
                                                'balance': {'mean', 'median'},
                                                'products': pd.Series.mode,
                                                'credit_card': pd.Series.mode,
                                                'last_activity': pd.Series.mode,
                                                'est_salary': {'mean', 'median'}
                                                }).T )
typical_user['Разница между портретами'] = typical_user[0] - typical_user[1]
typical_user['Разница между портретами в %'] = round(abs(typical_user['Разница между портретами']*100/typical_user[1]))
typical_user = typical_user.rename(columns = {0: 'Лояльные', 1 : 'Отточные'})
typical_user
```

Отдельно посчитаем моду по столбцам `city` и `gender`

```python
typical_user_2 = banks.groupby(['churn']).agg({ 'city': pd.Series.mode,
                                                'gender': pd.Series.mode
                                              }).T
typical_user_2 = typical_user_2.rename(columns = {0: 'Лояльные', 1 : 'Отточные'})
typical_user_2
```

Для начала создадим два новых датасета - в первый датасет `banks_churn_1` войдут все отточные пользователи, а во второй датасет `banks_churn_0` войдет все лояльные пользователи.

```python
banks_churn_0 = banks.query('churn == 0')
display(banks_churn_0.head())
banks_churn_0.info()
```

```python
banks_churn_1 = banks.query('churn == 1')
display(banks_churn_1.head())
banks_churn_1.info()
```

### Гипотеза №1 - Различие дохода между теми клиентами, которые ушли и теми, которые остались

Проверим гипотезу различия дохода между теми клиентами, которые ушли и теми, которые остались

Для начала сформируем нулевую (Но) и альтернативную (Н1) гипотезы:
* Hо - доходы отточных и лояльных клиентов равны;
* H1 - доходы отточных и лояльных клиентов разные.

В теории, сформулированные гипотезы можно проверить двумя способами - t-критерий Стьюдента (t-тест) или U-критерий Манна-Уитни. Рассмотрим каждый их них отдельно.

Чтобы понять, можно ли применять t-тест, нужно проверить три условия:
1. Генеральные совокупности не зависят друг от друга.\
В нашем случае имеется два отедльных, не зависящих друг от друга датасета `banks_churn_0` и `banks_churn_1`.
2. Выборочные средние нормально распределены.

```python
plt.subplots(figsize=(18, 8))

sns.distplot(banks_churn_0['est_salary'], label = 'Лояльные')
sns.distplot(banks_churn_1['est_salary'], label = 'Отточные')

plt.title('Распределение по est_salary', fontsize = 15)
plt.xlabel('', fontsize = 15)
plt.ylabel('', fontsize = 15)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.grid(True);
plt.legend(fontsize = 15)
plt.show();
```

Из графика видно, что значения дохода в столбце `est_salary` для отточных и лояльных клиентов не имеют нормального распределения.

```python
print('Средний доход лояльных клиентов равен:', banks_churn_0['est_salary'].mean())
print('Средний доход отточных клиентов равен:', banks_churn_1['est_salary'].mean())
print('Разница между средним доходом составляет:',  
      (banks_churn_1['est_salary'].mean()-banks_churn_0['est_salary'].mean())*100 / banks_churn_0['est_salary'].mean(), '%')
```

3. Дисперсии рассматриваемых генеральных совокупностей равны.

```python
print('Дисперсия лояльных клиентов равна:', banks_churn_0['est_salary'].var())
print('Дисперсия отточных клиентов равна:', banks_churn_1['est_salary'].var())
print('Разница между дисперсиями составляет:',  
      (banks_churn_1['est_salary'].var()-banks_churn_0['est_salary'].var())*100 / banks_churn_0['est_salary'].var(), '%')
```

По результатам проверки трех условий было выявлено, что применение t-теста для проверки наших гипотез - невозможно. В дополнение к этому, наши датасеты имеют выбросы по столбцу `est_salary`, что так же делает невозможным применение t-теста.\
В данном случае, при наличии больших (по сравнению с нормальным распределением) выбросов, подойдет непараметрический тест Уилкоксона-Манна-Уитни.

```python
alpha = 0.05  # задаем критический уровень статистической значимости

results = st.mannwhitneyu(banks_churn_0['est_salary'], banks_churn_1['est_salary'])

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу - доходы не равны ')
else:
    print('Не получилось отвергнуть нулевую гипотезу - доходы равны') 
```

**Промежуточный вывод**  

По результатам проведенного непараметрического теста Уилкоксона-Манна-Уитни выяснили, что различия дохода между теми клиентами, которые ушли и теми, которые остались - имеются.\
Однако необходимо понимать, что проверка данной гипотезы была проведена с учетом выбросов в данных. Главное отличие непараметрических методов в том, что они работают с рангами, — номерами значений в упорядоченном ряду, — никак не учитывая сами значения. Поэтому к ним прибегают тогда, когда работа с самими значениями невозможна из-за выбросов, сильно сдвигающих параметрические результаты

### Гипотеза №2 - Различие возраста между теми клиентами, которые ушли и теми, которые остались

Проверим гипотезу различия возраста между теми клиентами, которые ушли и теми, которые остались.

Для начала сформируем нулевую (Но) и альтернативную (Н1) гипотезы:
* Hо - возраст отточных и лояльных клиентов равный;
* H1 - возраст отточных и лояльных клиентов разный.

В теории, сформулированные гипотезы можно проверить двумя способами - t-критерий Стьюдента (t-тест) или U-критерий Манна-Уитни. Рассмотрим каждый их них отдельно.

Чтобы понять, можно ли применять t-тест, нужно проверить три условия:
1. Генеральные совокупности не зависят друг от друга.\
В нашем случае имеется два отедльных, не зависящих друг от друга датасета `banks_churn_0` и `banks_churn_1`.
2. Выборочные средние нормально распределены.

```python
plt.subplots(figsize=(18, 8))

sns.distplot(banks_churn_0['age'], bins = 68,  label = 'Лояльные')
sns.distplot(banks_churn_1['age'], bins = 55, label = 'Отточные')

plt.title('Распределение по age', fontsize = 15)
plt.xlabel('', fontsize = 15)
plt.ylabel('', fontsize = 15)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.grid(True);
plt.legend(fontsize = 15)
plt.show();
```

Из графика видно, что значения возраста в столбце `age` для отточных и лояльных клиентов близки к нормальному распределению.

```python
print('Средний возраст лояльных клиентов равен:', banks_churn_0['age'].mean())
print('Средний возраст отточных клиентов равен:', banks_churn_1['age'].mean())
print('Разница между средним возрастом составляет:',  
      (banks_churn_1['age'].mean() - banks_churn_0['age'].mean())*100 / banks_churn_0['age'].mean(), '%')
```

3. Дисперсии рассматриваемых генеральных совокупностей равны.

```python
print('Дисперсия по возрасту для лояльных клиентов равна:', banks_churn_0['age'].var())
print('Дисперсия по возрасту для отточных клиентов равна:', banks_churn_1['age'].var())
print('Разница между дисперсиями составляет:',  
      (banks_churn_1['age'].var() - banks_churn_0['age'].var())*100 / banks_churn_0['age'].var(), '%')
```

По результатам проверки трех условий было выявлено, что применение t-теста для проверки наших гипотез - возможно, однако при проведении t-теста, необходимо указать параметр `equal_var = False`. В данном случае проверяется двусторонняя гипотеза, поэтому параметр `alternative` останется равным `two-sided` (это значение задается по-умолчанию).

```python
alpha = 0.05  # задаем критический уровень статистической значимости

results = st.ttest_ind(banks_churn_0['age'], banks_churn_1['age'], equal_var = False)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу - возраст разный')
else:
    print('Не получилось отвергнуть нулевую гипотезу - возраст равный') 
```

**Промежуточный вывод**  

По результатам проведенного t-теста выяснили, что различия возраста между теми клиентами, которые ушли и теми, которые остались - есть. не смотря на схожесть построенного графика, возраст разный и по полученным расчетам средний возраст отточных клиентов равен 41,5, а лояльных 43 года.

Однако принимая во вномание не идеальное распределение значений столбца `age` и разницу на 18 % между значениями дисперсии, а так же наличие больших единичных значений (н-р возраст 86 лет), проведем дополнительную проверку гипотезы с помощью непараметрического теста Уилкоксона-Манна-Уитни.

```python
alpha = 0.05  # задаем критический уровень статистической значимости

results = st.mannwhitneyu(banks_churn_0['age'], banks_churn_1['age'])

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу - возраст разный')
else:
    print('Не получилось отвергнуть нулевую гипотезу - возраст равный')  
```

**Промежуточный вывод**  

Дополнительная проверка с помощью непараметрического теста Уилкоксона-Манна-Уитни так же показала, что возраст отточных и лояльных клиентов - разный.
