`Reach` — охват пользователей по 10-балльной шкале;\
`Impact` — влияние на пользователей по 10-балльной шкале;\
`Confidence` — уверенность в гипотезе по 10-балльной шкале;\
`Efforts` — затраты ресурсов на проверку гипотезы по 10-балльной шкале. Чем больше значение `Efforts`, тем дороже проверка гипотезы.

### Фреймворк ICE для приоритизации гипотез

Применим фреймворк ICE для приоритизации гипотез. Отсортируем их по убыванию приоритета.

Фреймворк `ICE` рассчитывается по следующей формуле:
<img src="https://pictures.s3.yandex.net/resources/ice_score_2-27da3b9e-15cd-4f96-91a4-fefffd1d986e_1569198107.jpg" width=600 height=600 />

Рассчитаем `ICE` и добавим соответсвующий столбец в нашу таблицу. 

```python
hypothesis['ICE'] =  round((hypothesis['Impact'] * hypothesis['Confidence'] /hypothesis['Efforts']), 2)
```

Отсортируем полученную таблицу по убыванию приоритета

```python
hypothesis.sort_values(by='ICE', ascending=False)
```

Отсортировав таблицу по фреймворку `ICE` видим, что наибольший приоритет получила гипотеза под номером 8 - "Запустить акцию, дающую скидку на товар в день рождения". Она окажет большое влияние на пользователей и продукт при относительно небольших затратах на её тестирование.

### Фреймворк RICE для приоритизации гипотез
Применим фреймворк RICE для приоритизации гипотез. Отсортируем их по убыванию приоритета.

Фреймворк `RICE` рассчитывается по следующей формуле:
<img src="https://pictures.s3.yandex.net/resources/rice_score_1634305501.png" width=600 height=600 />

Рассчитаем `RICE` и добавим соответсвующий столбец в нашу таблицу. 

```python
hypothesis['RICE'] = (hypothesis['Reach'] * hypothesis['Impact'] * hypothesis['Confidence']) / hypothesis['Efforts']
```

Отсортируем полученную таблицу по убыванию приоритета

```python
hypothesis.sort_values(by='RICE', ascending=False)
```

### Укажем, как изменилась приоритизация гипотез при применении RICE вместо ICE.

```python
hypothesis.sort_values(by='ICE', ascending=False)
```

**Промежуточный вывод:**  

При применении `ICE` наиболее перспективными гипотезами являются пять гипотез под порядковыми номерами 8, 0, 7, 6, 2.\
При применении же `RICE`, топ-5 перспективных гипотез уже меняется и выглядит следующим образом: 7, 2, 0, 6, 8.\
Фактически, в топ-5 попали те же самые гипотезы. \
Отличие топ-5 гипотез при применении `RICE` вместо `ICE` заключается в том, что в первом случае учитывается параметр `Reach` — скольких пользователей затронет изменение, которое вы хотите внести. И из-за того, что гипотеза под номером 7 имеет максимальную оценку по данному параметру (10 из 10) - это дает ей большое "преимущество" и делает её безоговорочным лидером приоритезации гипотез при применении `RICE`.

```python
orders_B = orders[orders['group'] == 'B']
print('Количестов пользователей/количество строк группы B равно:', orders_B.value_counts().count() )
```

Далее, с помощью метода `merge`, создадим таблицу `orders_A_and_B`, объединив таблицы `orders_A` и `orders_B` по общему столбцу `visitorId`. В качестве аргумента `how` метода передадим значение `inner`, что означает, что будут собраны данные из внутренней области (которые есть в обоих датафреймах).

```python
orders_A_and_B = orders_A.merge(orders_B, on = 'visitorId', how = 'inner')

# для удобства оставим столбцы с индефикатером пользователя, индентификатором заказа и группой А/В-теста
orders_A_and_B = orders_A_and_B[['visitorId', 'transactionId_x', 'group_x', 'transactionId_y', 'group_y']]
orders_A_and_B.head()
```

```python
#orders['visitorId'].value_counts().count()
orders_A_and_B['visitorId'].value_counts().count()
```

58 пользователей попали в обе группы данного теста. Данный факт говорит об ошибках при поведении A/B-теста, так как пользователи не должны попадать в обе группы одновременно. 58 пользователей - это 5,6 % от общего числа пользователей и данных пользователей можно исключить из датафрейма.

```python
# Передадим в переменную delete список с идентификаторами повторяющихся пользователей
delete = orders_A_and_B['visitorId'].unique()
#delete
```

```python
# Создадим срез таблицы orders. В итоговую таблицу не войдут повторяющиеся пользователи
orders = orders.query('visitorId not in @delete')
#orders.head()
```

```python
# Чтобы убедиться в том, что удаление прошлок корректно - посмотрим на разницу в пользователях таблицу orders
# до удаления и после. Разница должна составить 58 пользователей
visitors_after = orders['visitorId'].value_counts().count()
print('Количество пользователей до удаления:', visitors_before)
print('Количество пользователей после удаления:', visitors_after)
print('Было удалено', visitors_before - visitors_after, 'пользователей')
```

Агрегируем значения. Вычислим максимальную дату и для группы тоже рассчитаем максимум. Найдём число уникальных ID заказов и пользователей. Подсчитаем сумму средних чеков. Применим методы к каждой строке датафрейма и отсортируем результаты по столбцам `date` и `group`: 

```python
ordersAggregated = datesGroups.apply(lambda x: orders[np.logical_and(orders['date'] <= x['date'],
                                                                     orders['group'] == x['group'])]\
                    .agg({'date' : 'max', 'group' : 'max', 'transactionId' : 'nunique', 
                          'visitorId' : 'nunique', 'revenue' : 'sum'}), axis=1).sort_values(by=['date','group'])
```

Аналогично получим агрегированные кумулятивные по дням данные о посетителях интернет-магазина: 

```python
visitorsAggregated = datesGroups.apply(lambda x: visitors[np.logical_and(visitors['date'] <= x['date'], 
                                                                         visitors['group'] == x['group'])]\
                                .agg({'date' : 'max', 'group' : 'max', 'visitors' : 'sum'}), axis=1)\
                                .sort_values(by=['date','group'])
```

Объединим обе таблицы в одну с понятными названиями столбцов: 

```python
# создаем массив уникальных пар значений дат и групп теста
datesGroups = orders[['date','group']].drop_duplicates()

# получаем агрегированные кумулятивные по дням данные о заказах 
ordersAggregated = (datesGroups
    .apply(lambda x: orders[np.logical_and(orders['date'] <= x['date'], orders['group'] == x['group'])]
    .agg({'date' : 'max', 'group' : 'max', 'transactionId' : 'nunique', 'visitorId' : 'nunique', 'revenue' : 'sum'}), axis=1)
    .sort_values(by=['date','group']))

# получаем агрегированные кумулятивные по дням данные о посетителях интернет-магазина 
visitorsAggregated = ( datesGroups
                     .apply(lambda x: visitors[np.logical_and(visitors['date'] <= x['date'], visitors['group'] == x['group'])]
                     .agg({'date' : 'max', 'group' : 'max', 'visitors' : 'sum'}), axis=1)
                     .sort_values(by=['date','group']) )

# объединяем кумулятивные данные в одной таблице и присваиваем ее столбцам понятные названия
cumulativeData = ordersAggregated.merge(visitorsAggregated, left_on=['date', 'group'], right_on=['date', 'group'])
cumulativeData.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']

# Выводим первые 5 строк созданной таблицы
cumulativeData.head(5)
```

Объявим переменные `sampleA` и `sampleB`, в которых пользователям из разных групп будет соответствовать количество заказов. Тем, кто ничего не заказал, будут соответствовать нули.

Дополнительно выведем относительный прирост среднего числа заказов группы B: среднее число заказов группы B / среднее число заказов группы A - 1.  Округлим до трёх знаков после запятой.

```python
sampleA = pd.concat(
    [
        ordersByUsersA['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])
            ), name='orders',
        ),
    ], axis=0,
)

sampleB = pd.concat(
    [
        ordersByUsersB['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)

print("p-value равно: {0:.3f}".format(st.mannwhitneyu(sampleA, sampleB)[1]))

print("Относительный прирост среднего числа заказов группы B равен: {0:.3%}".format(sampleB.mean() / sampleA.mean() - 1))
```

**Промежуточный вывод:**  

Полученное значение `p-value` = 0.011 и оно меньше статистической значимости 0.05. Значит отвергаем нулевую гипотезу о том, что различий в среднем количестве заказов между группами нет и принимаем альтернативную гипотезу о том, что различия в среднем между группами есть. \
Относительный прирост среднего числа заказов группы B равен 15.98 %.

### Статистическая значимость различий в среднем чеке заказа между группами по «сырым» данным

Теперь проверим статистическую значимость различий в среднем чеке между сегментами.
Сформулируем гипотезы:  
* Но - различий в среднем чеке между группами нет.
* Н1 - различия в среднем чеке между группами есть.

Чтобы рассчитать статистическую значимость различий в среднем чеке, передадим критерию `mannwhitneyu()` данные о выручке с заказов. 

```python
print('p-value равно: {0:.3f}'.format(st.mannwhitneyu(orders[orders['group']=='A']['revenue'],
                                                               orders[orders['group']=='B']['revenue'])[1]))
print('Относительный прирост среднего чека заказа группы B равен: {0:.2%}'.format(orders[orders['group']=='B']['revenue']
                                                                    .mean()/orders[orders['group']=='A']['revenue'].mean()-1))
```

**Промежуточный вывод:**  

Полученное значение p-value = 0.829 значительно больше 0.05. Значит принимаем нулевую гипотезу о том, что различий в среднем чеке между группами нет.\
Однако у нас зафиксирован относительный прирост среднего чека заказа группы B на 28,66 %. Такая большая разница получена за счет выбросов, имеющихся в группе В. ### Статистическая значимость различий в среднем количестве заказов на посетителя между группами по «очищенным» данным

Посчитаем статистическую значимость различий в среднем количестве заказов на посетителя между группами по «очищенным» данным.

Напомним, что за аномальных пользователей были приняты те пользователи, кто совершил больше 2 заказов или заказ дороже 26 785 (включительно).\
Сделаем срезы пользователей с числом заказов больше 2 — `usersWithManyOrders` и пользователей, совершивших заказы дороже 26 785 — `usersWithExpensiveOrders`. Объединим их в таблице `abnormalUsers`.

```python
#display(revenue_lim)
#orders_lim
```

```python
usersWithManyOrders = pd.concat(
    [
        ordersByUsersA[ordersByUsersA['orders'] > orders_lim]['visitorId'],
        ordersByUsersB[ordersByUsersB['orders'] > orders_lim]['visitorId'],
    ],
    axis=0,
)
usersWithExpensiveOrders = orders[orders['revenue'] >= revenue_lim]['visitorId']
abnormalUsers = (
    pd.concat([usersWithManyOrders, usersWithExpensiveOrders], axis=0)
    .drop_duplicates()
    .sort_values()
)
print(abnormalUsers.head(5))
print()
print('Количество пользователей:', abnormalUsers.shape[0])
```

Узнаем, как их действия повлияли на результаты теста. Посчитаем статистическую значимость различий в среднем количестве заказов между группами теста по очищенным данным. Сформулированные гипотезы остались те же:
* Но - различий в среднем количестве заказов между группами нет.
* Н1 - различия в среднем между группами есть.

Сначала подготовим выборки количества заказов по пользователям по группам теста:

```python
sampleAFiltered = pd.concat(
    [
        ordersByUsersA[
            np.logical_not(ordersByUsersA['visitorId'].isin(abnormalUsers))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)

sampleBFiltered = pd.concat(
    [
        ordersByUsersB[
            np.logical_not(ordersByUsersB['visitorId'].isin(abnormalUsers))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)
print('p-value равно: {0:.3f}'.format(st.mannwhitneyu(sampleAFiltered, sampleBFiltered)[1]))
print('Относительный прирост среднего числа заказов группы B равен: {0:.3%}'.format(sampleBFiltered.mean()/sampleAFiltered.mean()-1))
```

### Статистическая значимость различий в среднем чеке заказа между группами по «очищенным» данным

* Но - различий в среднем чеке между группами нет.
* Н1 - различия в среднем чеке между группами есть.

```python
print(
    'p-value равно: {0:.3f}'.format(
        st.mannwhitneyu(
            orders[
                np.logical_and(
                    orders['group'] == 'A',
                    np.logical_not(orders['visitorId'].isin(abnormalUsers)),
                )
            ]['revenue'],
            orders[
                np.logical_and(
                    orders['group'] == 'B',
                    np.logical_not(orders['visitorId'].isin(abnormalUsers)),
                )
            ]['revenue'],
        )[1]
    )
)

print(
    "Относительный прирост среднего чека заказа группы B равен: {0:.2%}".format(
        orders[
            np.logical_and(
                orders['group'] == 'B',
                np.logical_not(orders['visitorId'].isin(abnormalUsers)),
            )
        ]['revenue'].mean()
        / orders[
            np.logical_and(
                orders['group'] == 'A',
                np.logical_not(orders['visitorId'].isin(abnormalUsers)),
            )
        ]['revenue'].mean()
        - 1
    )
)
```
